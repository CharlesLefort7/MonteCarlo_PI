```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np
from random import seed
from random import random
import matplotlib.pyplot as plt



seed(1)


ite = 1000000 # Nombre de réalisations d'histoires
dans_lac = 0 # Nombre de fois que l'on se trouve dans le cercle




histox = np.ones((ite))
histoy = np.ones((ite))

histoxBIS = np.ones((ite))
histoyBIS = np.ones((ite))

pi = np.zeros((ite))
var = np.zeros((ite))

for i in range(ite):
    
    x = random()
    y = random()
    
    r = np.sqrt(x**2+y**2)
    
    if r < 1 :
        dans_lac = dans_lac + 1
        histox[i] = x
        histoy[i] = y
    else:
        histoxBIS[i] = x
        histoyBIS[i] = y
        
    pi [i] = 4 * dans_lac / ite
    var[i] = (1-pi[i]) * np.square(-pi[i]) + (pi[i]/4) * np.square(4- pi[i])
    


print("Pi = %f" % pi[ite-1])
print("Var = %f" % var[ite-1])



plt.figure()
plt.plot(histox,histoy,'ro')
plt.plot(histoxBIS,histoyBIS,'bs')

plt.figure()
plt.plot(pi)
plt.figure()
plt.plot(var)
```
